```python
from sklearn.neural_network import MLPRegressor, MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.metrics import accuracy_score
import pandas as pd 
import numpy as np
```

```python
#importing the file as a Data Frame
file_name = 'GOT_DPE.xlsx'
df = pd.read_excel(file_name)
df.info()
```

```python
#Data pre processing
got = df
got = got.drop(['index', 'S.No', 'plod', 'name', 'title', 'dateOfBirth', 'DateoFdeath', 
               'mother', 'father', 'heir', 'spouse', 'isAliveMother', 'isAliveFather', 
               'isAliveHeir', 'isAliveSpouse', 'isNoble', 'age', 'isPopular'], axis=1)
got.info()
```

```python
# Replacin NAN's with a Value
got.fillna('Unaffiliated', inplace=True)
got.info()
```

```python
# One hot Encoding the Categorial data of House and Culture
#Initializing the One hot Encoder
ohe = OneHotEncoder()
#Declaring the Columns we want to encode
categorical_cols = ['culture', 'house']
#Creating the array with the encoded Values
ft_array = ohe.fit_transform(got[categorical_cols]).toarray()
#Getting the labels for the category
ft_labels = ohe.categories_
ft_labels = np.array(ft_labels).ravel()
#Concatenating the 2 np arrays within labels
ft_labels_combined = np.concatenate(ft_labels)
#Dataframe with encoded values
encoded_df = pd.DataFrame(ft_array, columns = ft_labels_combined)
encoded_df.info()
encoded_df.head()
```

```python
#Splitting into X and Y data 
y = got['isAlive']  # Selecting the 'isAlive' column as the target variable
# Dropping the 'isAlive' column to get the features and dropping the unecoded columns
x = got.drop(['isAlive', 'culture', 'house'], axis=1) 
x = pd.concat([x, encoded_df], axis=1) #Concatenating the encoded df to the x dataframe
```

```python
# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

# Create an MLPClassifier instance
death_classifier = MLPClassifier(hidden_layer_sizes=(12, 24, 12), max_iter=4000, random_state=42, activation='relu',
                                verbose=True, tol=1e-4, n_iter_no_change=50, learning_rate_init=0.0001)

# Fit the model to the training data
death_classifier.fit(x_train, y_train)

# Make predictions on the testing set
y_pred = death_classifier.predict(x_test)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
```

```python
# Standardize the features
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

# Create a KNeighborsClassifier instance
knn_classifier = KNeighborsClassifier(n_neighbors=50)  

# Fit the model to the training data
knn_classifier.fit(x_train_scaled, y_train)

# Make predictions on the testing set
y_pred = knn_classifier.predict(x_test_scaled)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
```

```python
# Create a RandomForestClassifier instance
random_forest_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Fit the model to the training data
random_forest_classifier.fit(x_train, y_train)

# Make predictions on the testing set
y_pred = random_forest_classifier.predict(x_test)

# Evaluate the accuracy
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy}')
```
